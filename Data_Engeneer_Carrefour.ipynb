Criando schema para validação da coluna codigo_ocorrencia  
**schmas permite a validação em tempo real e apresenta erro caso haja inconsistências de dados.**

Alterando a quantidade de asteriscos da coluna ocorrencia_uf no indice 2227 de 3 para 2 asteriscos

```python
# O metodo Check.str_matches valida a hora através de uma expressão regular python
schema = pa.DataFrameSchema(
    columns = {
        #'codigo':pa.Column(pa.Int, required=False), # required=False -permite a criação de uma coluna que não existe no dataframe
        'codigo_ocorrencia':pa.Column(pa.Int),
        'codigo_ocorrencia2':pa.Column(pa.Int),
        'ocorrencia_classificacao':pa.Column(pa.String),
        'ocorrencia_cidade':pa.Column(pa.String),
        'ocorrencia_uf':pa.Column(pa.String, pa.Check.str_length(2, 2)),
        'ocorrencia_aerodromo':pa.Column(pa.String),
        'ocorrencia_dia': pa.Column(pa.DateTime),
        'ocorrencia_hora':pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$') ,nullable=True),# nullable=True - permite a validação de campus null
        'total_recomendacoes':pa.Column(pa.Int)
    }
)
```

validando schema

```python
schema.validate(df_ocorrencia)
```

Erro identificado depois da validação no schema

**df_ocorrencia.isnull** mostra dados não informado  
**df_ocorrencia.fillna(0)** altera os valores onde ha null para 0
